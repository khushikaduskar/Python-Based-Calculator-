```python
import tkinter as tk

```

```python
# This variable stores what user types (numbers and operators)
expression = ""

# This function runs when any button is pressed
def press(key):
    global expression

    # If user presses "=" → calculate result
    if key == "=":
        # Replace calculator symbols with Python symbols
        expression = str(eval(expression.replace("×", "*").replace("÷", "/")))

    # If user presses "C" → clear everything
    elif key == "C":
        expression = ""

    # Otherwise add number or operator to expression
    else:
        expression += key

    # Show expression on calculator screen
    display.set(expression if expression else "0")


# Create main window
root = tk.Tk()
root.title("Calculator")
root.geometry("300x400")
root.resizable(False, False)

# Variable to show text on display
display = tk.StringVar(value="0")

# Calculator display screen
screen = tk.Label(
    root,
    textvariable=display,
    font=("Arial", 24),
    bg="black",
    fg="white",
    anchor="e",
    padx=10
)
screen.pack(fill="x")

# Calculator buttons
buttons = [
    "7","8","9","÷",
    "4","5","6","×",
    "1","2","3","-",
    "0",".","=","+"
]

# Frame for buttons
frame = tk.Frame(root)
frame.pack(expand=True, fill="both")

# Clear button
tk.Button(
    frame,
    text="C",
    font=("Arial", 16),
    command=lambda: press("C")
).grid(row=0, column=0, columnspan=4, sticky="nsew")

# Add number and operator buttons
row = 1
col = 0
for b in buttons:
    tk.Button(
        frame,
        text=b,
        font=("Arial", 16),
        command=lambda x=b: press(x)
    ).grid(row=row, column=col, sticky="nsew")

    col += 1
    if col == 4:
        col = 0
        row += 1

# Make buttons adjust to window size
for i in range(5):
    frame.rowconfigure(i, weight=1)
for j in range(4):
    frame.columnconfigure(j, weight=1)

# Start the calculator
root.mainloop()
```
